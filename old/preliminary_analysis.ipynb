```python
new_df = df[["name","star_name","star_metallicity","mass","radius"]].copy()
```

```python
new_df.info()
```

```python
threshold = 3

# Calculate the density
# Density = mass / volume
new_df['density'] = new_df['mass'] * (1.899e+30) / (4/3 * np.pi * new_df['radius']**3 * (7.149e+9)**3 )

# Add the 'rocky' column based on the density threshold
new_df['rocky'] = (new_df['mass'] < (10 * 0.00314558) ).astype(int)

# Print the updated DataFrame
print(new_df.head())
```

```python
c = 0
for _,i in new_df.iterrows():
    if i['rocky'] == 1 and i['density'] > threshold:
        print(f"{i['name']}: radius = {i['radius']} mass = {i['mass']} density = {i['density']} rocky = {i['rocky']}")
        c += 1
print(f"number of rocky planets {c}")
```

```python
value_counts = new_df["star_name"].value_counts()

# Accessing the index (unique values)
unique_values = value_counts.index.tolist() #or .to_numpy()

# Accessing the values (counts)
counts = value_counts.values.tolist() #or .to_numpy()

print("Unique Values:", unique_values)
print("Counts:", counts)

# #or print them together.
# for value, count in value_counts.items():
#     print(f"Value: {value}, Count: {count}")
```

```python
for name in unique_values[:10]:
    for _,i in new_df.iterrows():
        if i["star_name"] == name:
            print(name)
            print(i["star_metallicity"])
            break
```

```python
# Group the DataFrame by 'star_name'
grouped_by_star = new_df.groupby('star_name')

# Calculate the number of planets for each star
planet_counts = grouped_by_star['name'].count().rename('planet_count')

# Get the unique star metallicity for each star
# We can take the first metallicity value within each group, assuming it's consistent for a given star
star_metallicities = grouped_by_star['star_metallicity'].first()

# Combine the planet counts and star metallicities
star_planet_info = pd.concat([planet_counts, star_metallicities], axis=1).reset_index()

# Print the resulting DataFrame
print(star_planet_info)
```

```python
# Sort the DataFrame based on the 'planet_count' column in descending order
sorted_by_planet_count_desc = star_planet_info.sort_values(by='planet_count', ascending=False)
print(sorted_by_planet_count_desc.columns)

for _,i in sorted_by_planet_count_desc[:100].iterrows():
    print(f"{i['star_name']}: number of planets {i['planet_count']}: metallicity {i['star_metallicity']}")
    
```
